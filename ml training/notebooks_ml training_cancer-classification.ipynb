```python
import seaborn as sns

labels = [['True Neg','False Pos'],['False Neg','True Pos']]
df_cm = pd.DataFrame(conf_matrix)

cm_sum = np.sum(df_cm, axis=1)
cm_perc = df_cm / cm_sum.astype(float) * 100
annot = np.empty_like(df_cm).astype(str)

nrows, ncols = df_cm.shape
for i in range(nrows):
    for j in range(ncols):
        c = df_cm[i][ j]
        p = cm_perc[i][ j]
        if i == j:
            s = cm_sum[i]
            annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
        elif c == 0:
            annot[i, j] = ''
        else:
            annot[i, j] = '%.1f%%\n%d' % (p, c)
                

sns.set(font_scale=1)
ax = sns.heatmap(df_cm, annot=annot, fmt='', cmap='Blues')

ax.set_xlabel("Predicted type", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['Negative', 'Positive'])

# set y-axis label and ticks
ax.set_ylabel("Actual type", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['Negative', 'Positive'])

# set plot title
ax.set_title("Confusion Matrix for the Cancer cells detection Model", fontsize=14, pad=20)

```
